```python
prev_year_mask = df["data fatturazione prevista"].dt.year < current_year
curr_year_mask = df["data fatturazione prevista"].dt.year == current_year
next_year_mask = df["data fatturazione prevista"].dt.year > current_year
```

```python
from util.calendar import int_month_to_str

currDF = df[curr_year_mask]
currDF["data fatturazione prevista"] =  (currDF["data fatturazione prevista"].dt.month).apply(int_month_to_str)
```

```python
for m in range(1, 13):
    month = int_month_to_str(m)
    month_mask = currDF["data fatturazione prevista"] == month
    currDF[month] = np.where(currDF["data fatturazione prevista"] == month, currDF["importo acconto fatturato"], 0)
    
currMonthly = currDF.groupby(currDF.index).sum()
```
