```python
import pandas as pd
import numpy as np
from util import calendar
```

```python
src = pd.read_excel("cashflow.xlsx", parse_dates=[1], engine="openpyxl", names=["code","date","amount"])
df = src.copy()
df.head(10)
```

```python
df["m"], df["y"] = (df["date"].dt.month).apply(calendar.int_month_to_str), df["date"].dt.year
df.head(10)
```

```python
current_year = 2022
df["prev"] = np.where(df["y"] < current_year, df["amount"], 0)
df["next"] = np.where(df["y"] > current_year, df["amount"], 0)
df.head(10)
```

```python
for _, m in calendar.months.items():
    df[m] = np.where((df["m"] == m) & (df["y"] == current_year), df["amount"], 0)
df
```

```python
agg = df.groupby(df["code"]).sum().sort_index()
agg = agg.drop("y", axis=1)
agg = agg.rename(columns={"prev":"Anni Precedenti", "next":"Anni Successivi", "amount":"Totale Fatturato"})
```

```python
agg.head(10)
```

```python
agg2 = agg.copy()
newMonthCols = {}
for _, m in calendar.months.items(): newMonthCols[m] = m + " Fatturato"
agg2 = agg2.rename(columns=newMonthCols)
agg2.head(10)
```

```python
for _, m in calendar.months.items():
    agg2[m + " Incasso"] = 0
agg2["Totale Incasso"] = 0
agg2.head(10)
```

```python
newColOrder1 = [1] + [m + 3 for m in range(0, 12)] + [2, 0]
newColOrder2 = [1] + np.array([(m + 3, m + 3 + 12) for m in range(12)]).flatten().tolist() + [2, 0, -1]
final1 = agg.iloc[:,np.array(newColOrder1)]
final2 = agg2.iloc[:,np.array(newColOrder2)]
```
